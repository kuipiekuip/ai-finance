```python
from crewai import Agent, Task, Crew, LLM

llm_groq = LLM(model="groq/llama3-8b-8192")
# Define specialized agents
revenue_agent = Agent(
    role="Revenue Extraction Specialist",
    goal="Extract and return all revenue-related numbers from financial reports in JSON format.",
    backstory="An expert in financial data analysis, specialized in identifying revenue growth trends.",
    allow_delegation=False,
    verbose=True,
    llm=llm_groq
)

profit_agent = Agent(
    role="Profit Extraction Specialist",
    goal="Extract net income, operating income, and profit margins from financial statements.",
    backstory="A financial analyst focused on corporate profitability.",
    allow_delegation=False,
    verbose=True,
    llm=llm_groq
)

cash_flow_agent = Agent(
    role="Cash Flow Extraction Specialist",
    goal="Extract cash flow metrics, including operating cash flow and free cash flow.",
    backstory="Expert in liquidity and cash flow analysis.",
    allow_delegation=False,
    verbose=True,
    llm=llm_groq
)
```

```python
revenue_task = Task(
    description=(
        "Extract revenue figures, including quarterly and yearly values, "
        "year-over-year (YoY) and quarter-over-quarter (QoQ) growth, and any revenue segments.\n"
        "Return the data strictly in the following JSON format:\n"
        '{\n  "Revenue": {\n    "Q4_2024": {"Amount": "X", "YoY_Growth": "X%"},\n'
        '    "FY_2024": {"Amount": "X", "YoY_Growth": "X%"},\n'
        '    "Segments": {"U.S_Commercial": "X", "Government": "X"}\n  }\n}'
    ),
    agent=revenue_agent,
    expected_output="Valid JSON format with revenue figures"
)

profit_task = Task(
    description=(
        "Extract net income, operating income, and profit margins. "
        "Return the data in this structured JSON format:\n"
        '{\n  "Profitability": {\n    "Net_Income": {\n'
        '      "Q4_2024": {"Amount": "X", "Net_Margin": "X%"},\n'
        '      "FY_2024": {"Amount": "X", "Net_Margin": "X%"}\n    }\n  }\n}'
    ),
    agent=profit_agent,
    expected_output="Valid JSON format with profit figures"
)

cash_flow_task = Task(
    description=(
        "Extract cash flow data, including operating cash flow and free cash flow. "
        "Return the data in this structured JSON format:\n"
        '{\n  "Cash_Flow": {\n    "Operating_Cash_Flow": {\n'
        '      "Q4_2024": "X", "FY_2024": "X"\n    },\n'
        '    "Free_Cash_Flow": {\n      "Q4_2024": "X", "FY_2024": "X"\n    }\n  }\n}'
    ),
    agent=cash_flow_agent,
    expected_output="Valid JSON format with cash flow figures"
)
```

```python
import json
# Create Crew with dependencies
crew = Crew(
    agents=[
        revenue_agent, profit_agent, cash_flow_agent,
    ],
    tasks=[
        revenue_task, profit_task, cash_flow_task,
    ]
)

# Run the pipeline
results = crew.kickoff()

# Convert output to JSON
final_json = json.dumps(results, indent=2)
print(final_json)
```
